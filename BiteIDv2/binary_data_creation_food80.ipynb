```python
multiclass_train_dir = Path("food80_dataset/train")

class_names = [name for name in os.listdir(multiclass_train_dir)]
```

```python
binary_class_names = ["0_non-food", "1_food"]

try:
  binary_dir = Path("data/food80_binary")
  os.makedirs(binary_dir)
  splits = ["train", "test"]

  for split in splits:
    split_path = os.path.join(binary_dir, split)

    os.mkdir(split_path)

    for class_name in binary_class_names:
      class_dir = os.path.join(split_path, class_name)

      os.mkdir(class_dir)


except:
  print("paths already created")
```

### Getting all the image_paths from the train directory of the multiclass dataset

Plan is to get all of them and randomly pick 1125 of them to be in the '1_food' directory of the train folder of the food80_binary dataset

```python

import glob
import shutil

# Grab the image_paths
train_image_paths = []

for class_name in class_names:
  class_dir = multiclass_train_dir / class_name
  image_paths = glob.glob(f"{class_dir}/*.jpg")

  for path in image_paths:
    train_image_paths.append(path)
```

```python
# Randomly pick 1125 images. Don't need to worry about shuffling now. Making dataloaders will do it

import random

chosen_train_images = random.sample(train_image_paths, k=1125)
```

```python
# Copy images from the file paths in 'chosen_train_images'

for image_path in chosen_train_images:
  shutil.copy(image_path, "data/food80_binary/train/1_food")


## the food folder for train should have 1125 images now. Remember: 1125 images for each train class (food, non-food), 375 images for each test class (food, non-food)
```

### Getting all the image_paths from the test directory of the multiclass dataset

Plan is to get all of them and randomly pick 375 of them to be in the '1_food' directory of the test folder of the food80_binary dataset

```python
multiclass_test_dir = Path("food80_dataset/test")
# Grab the image_paths
test_image_paths = []

for class_name in class_names:
  class_dir = multiclass_test_dir / class_name
  image_paths = glob.glob(f"{class_dir}/*.jpg")

  for path in image_paths:
    test_image_paths.append(path)
```

```python
import random

chosen_test_images = random.sample(train_image_paths, k=375)
```

```python
for image_path in chosen_test_images:
  shutil.copy(image_path, "data/food80_binary/test/1_food")

# the food folder for test should have 375 images now.
```

### Now I need to split the images from 'binary_data_nonfood.zip" for 1125 images (train) and 375 images (test)

Upload the zip file manually

```python
import zipfile
from zipfile import ZipFile

with zipfile.ZipFile("binary_data_nonfood.zip", "r") as myzip:
  myzip.extractall()


nonfood_image_paths = []

for image_id in os.listdir("binary_data/nonfood"):
  image_path = os.path.join("binary_data/nonfood", image_id)
  nonfood_image_paths.append(image_path)


train_nonfood_split = nonfood_image_paths[:1125]
test_nonfood_split = nonfood_image_paths[1125:]

```

```python
# Copy images from the splits to the binary dataset

binary_train_nonfood_dir = binary_dir / "train" / "0_non-food"
binary_test_nonfood_dir = binary_dir / "test" / "0_non-food"

for image_path in train_nonfood_split:
  shutil.copy(image_path, binary_train_nonfood_dir)

for image_path in test_nonfood_split:
  shutil.copy(image_path, binary_test_nonfood_dir)
```

```python
# Don't want the 'data' dir anymore

# Path to the 'data' directory
data_dir = "data"

# Move contents of 'data' to the parent directory
for item in os.listdir(data_dir):
    shutil.move(os.path.join(data_dir, item), '.')

# Remove the empty 'data' directory
os.rmdir(data_dir)
```
